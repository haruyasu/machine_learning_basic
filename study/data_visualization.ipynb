# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0, 10)
```

```python
y = x ** 2
```

```python
plt.plot(x, y, 'red')
plt.xlim(0, 4)
plt.ylim(0, 10)
plt.title('TITLE')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
```

```python
mat = np.arange(0, 100).reshape(10, 10)
```

```python
plt.imshow(mat, cmap='coolwarm')
```

```python
mat = np.random.randint(0, 1000, (10, 10))
```

```python
plt.imshow(mat)
plt.colorbar()
```
