```python
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-12, 12, 36)
y = np.linspace(-12, 12, 36)

x,y = np.meshgrid(x, y)
r = np.sqrt(x**4 + y**4)
z = np.cos(r) * np.sin(r)


fig = plt.figure()
ax = plt.axes(projection="3d")
ax.contour3D(x, y, z, 30)
```

```python
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# データ作成
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# Surface Plot
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.random.rand(2, 100) * 4
hist, xedges, yedges = np.histogram2d(x, y, bins=4, range=[[0, 4], [0, 4]])


# np.meshgridは（ny、nx）の配列を与え、xposを平坦化するために 'F'を使用
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25)

"""
xposの中身
[[ 0.25  1.25  2.25  3.25]
 [ 0.25  1.25  2.25  3.25]
 [ 0.25  1.25  2.25  3.25]
 [ 0.25  1.25  2.25  3.25]]

yposの中身
[[ 0.25  0.25  0.25  0.25]
 [ 1.25  1.25  1.25  1.25]
 [ 2.25  2.25  2.25  2.25]
 [ 3.25  3.25  3.25  3.25]]
"""

xpos = xpos.flatten('F')
ypos = ypos.flatten('F')

"""
上記処理によって、xposとyposの中身は以下のようになります

xposの中身
[ 0.25  0.25  0.25  0.25  1.25  1.25  1.25  1.25  2.25  2.25  2.25  2.25
  3.25  3.25  3.25  3.25]

yposの中身
[ 0.25  1.25  2.25  3.25  0.25  1.25  2.25  3.25  0.25  1.25  2.25  3.25
  0.25  1.25  2.25  3.25]
"""

zpos = np.zeros_like(xpos)

dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
# 次のように、colorやzsortなどのオプションを指定することも可能です。
# ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')
```
